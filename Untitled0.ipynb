```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1. Set paths for training and testing data
train_dir = '/content/emotion/train'  # Path to the training data directory
test_dir = '/content/emotion/test'    # Path to the testing data directory

# 2. Define parameters for image size and batch size
IMG_SIZE = 100  # Assuming the image size is 100x100 like in the age dataset
BATCH_SIZE = 32  # Batch size for training

# 3. Create ImageDataGenerator for data preprocessing and augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,        # Normalize pixel values (0-255 to 0-1)
    rotation_range=10,        # Data augmentation: random rotation
    width_shift_range=0.1,    # Data augmentation: horizontal shift
    height_shift_range=0.1,   # Data augmentation: vertical shift
    zoom_range=0.1,           # Data augmentation: random zoom
    horizontal_flip=True,     # Data augmentation: random horizontal flip
)

# 4. Use a separate ImageDataGenerator for the test set without augmentation, with validation split
test_datagen = ImageDataGenerator(
    rescale=1.0/255.0,        # Normalize pixel values (0-255 to 0-1)
    validation_split=0.5      # Split test data into 50% validation, 50% test
)

# 5. Load and preprocess the training and validation data from the train folder
emotion_train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),  # Resize images to (100, 100)
    batch_size=BATCH_SIZE,
    class_mode='categorical',          # For multi-class classification
    subset='training'                  # Use 80% of images for training
)



# 6. Load and preprocess the validation and test data from the test folder
emotion_validation_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'                # Use 50% of the test folder for validation
)

# Test set (remaining 50% of the test folder)
emotion_test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'                  # Use the other 50% of the test folder for testing
)

# 7. Check the class labels assigned by ImageDataGenerator
print("Class indices:", emotion_train_generator.class_indices)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, GlobalAveragePooling2D,Dropout
```

```python
base_model = MobileNetV2(input_shape=(100, 100, 3), include_top=False, weights='imagenet')
```

```python
base_model.trainable = False
x=base_model.output
x=GlobalAveragePooling2D()(x)
x=Dropout(0.2)(x)
emotion_output = Dense(5, activation='softmax', name='emotion_output')(x)
age_output = Dense(1, activation='linear', name='age_output')(x)
```

```python
model1=Model(inputs=base_model.input,outputs=age_output)
model2=Model(inputs=base_model.input,outputs=emotion_output)
```

```python
model1.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
